```python
import numpy as np
import pandas as pd
```

```python
dict= np.load('tesla.npy',allow_pickle='TRUE').item()
df = pd.DataFrame.from_dict(dict,orient='index')
df
```

```python
df.columns = ['Open', 'High', 'Low','Close','Adjusted Close','Volume','Dividend Amount','Split Coefficinet']
df
```

```python
# change index  to datatime format
df.index =pd.to_datetime(df.index)

# change other columns to numerical format
df=df.astype('float') 
```

```python
# save dataframe to csv
df.to_csv('Tesla.csv')
```

```python
# read csv to dataframe
df_tesla = pd.read_csv('Tesla.csv')

df_tesla.info()
```

```python
# rename name of column
df_tesla.rename(columns = {'Unnamed: 0':'Date'}, inplace = True)

df_tesla.info()
```

```python
# change column to Date format
df_tesla['Date'] =  pd.to_datetime(df_tesla['Date'])
```

```python
# save dataframe to csv
df_tesla.to_csv('Tesla_final.csv')
```

### Data visualisation

https://plotly.com/python/candlestick-charts/

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime



fig = go.Figure(data=[go.Candlestick(x=df_tesla['Date'],
                open=df_tesla['Open'],
                high=df_tesla['High'],
                low=df_tesla['Low'],
                close=df_tesla['Close'])])

fig.show()
```

```python
from plotly.subplots import make_subplots
fig =make_subplots(specs=[[{'secondary_y':True}]])

fig.add_trace(go.Candlestick(x=df_tesla['Date'],
                open=df_tesla['Open'],
                high=df_tesla['High'],
                low=df_tesla['Low'],
                close=df_tesla['Close']),secondary_y=True)

# include a go.Bar trace for volumes
fig.add_trace(go.Bar(x=df_tesla['Date'], y=df_tesla['Volume'],marker_color="pink",opacity=0.5),
               secondary_y=False,)

fig.update_layout(showlegend=False)

fig.layout.yaxis2.showgrid=False
fig.show()

```

### Update chart (example)

```python
import plotly.graph_objects as go

colors = ['rgb(158,202,225)',] * 3
colors[0] = 'crimson'
 

x = ['Product A', 'Product B', 'Product C']
y = [20, 14, 23]

# Use the hovertext kw argument for hover text
fig = go.Figure(data=[go.Bar(x=x, y=y,
            hovertext=['27% market share', '24% market share', '19% market share'], 
            # text=['27% market share', '24% market share', '19% market share'],
            # textposition='inside')
)])

# Customize aspect
fig.update_traces(
                  # marker_color='rgb(158,202,225)',
                  marker_line_color='rgb(8,48,107)',
                  marker_color = colors,
                  marker_line_width=1.5, opacity=0.6)

fig.update_layout(title_text='January 2013 Sales Report')

fig.show()
```
